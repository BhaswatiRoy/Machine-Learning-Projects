```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#g=green,r=red,c=cyan,k=black,y=yellow,m=magenta,w=white,b=blue
```

```python
#Importing the dataset
df=pd.read_csv("Father Son Height.csv")
```

```python
#giving new shape to array to a compatible shape for processings without changing data
#reshape(-1,1)=>means to reshape the arrray to have only 1 column and as many rows as needed
#assigning values of each column to variables x and y
x=df["Father"].values.reshape(-1,1)
y=df["Son"].values
```

```python
#importing Linear Regression Model
from sklearn.linear_model import LinearRegression
#assigning the function to variable lin_reg
lin_reg=LinearRegression()
#fitting parameters x,y into the model
lin_reg.fit(x,y)
```

```python
#plotting linear plot
plt.plot(x,y)
plt.title("Linear Plot")
plt.xlabel("Father's Height")
plt.ylabel("Son's Height")
plt.grid(True,color="k")
plt.show()
```

```python
#plotting scatter plot
plt.scatter(x,y,color="magenta")
plt.title("Scatter Plot")
#plotting best fit line where lin_reg.predict(x)=y coordinate
plt.plot(x,lin_reg.predict(x),color="black",linewidth="2")
plt.xlabel("Father's Height")
plt.ylabel("Son's Height")
plt.grid(True,color="k")
plt.show()
```

```python
#importing polynomial features incase if the relation of best fit line is not a linear curve
from sklearn.preprocessing import PolynomialFeatures
#assigning the function to a variable poly_reg with a degree of 4
poly=PolynomialFeatures(degree=4)
#fit_transform=>scaling the training data and also scaling the parameters of the training data
x_poly=poly.fit_transform(x)
lin_reg_poly=LinearRegression()
#fitting parameters x_poly,y_poly into the model
lin_reg_poly.fit(x_poly,y)
```

```python
#plotting the scatter plot
plt.scatter(x,y,color="red")
plt.title("Scatter Plot")
#plotting the best fit line
plt.plot(x,lin_reg_poly.predict(poly.fit_transform(x)),color="black",linewidth="1")
plt.xlabel("Father's Age")
plt.ylabel("Son's Age")
plt.grid(True,color="w")
```
